```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
df = sns.load_dataset('titanic')
```

```python
df.describe(include='all')
```

```python
df.columns
```

```python
df.info()
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
```

```python
sns.countplot(x='pclass', hue='survived', data=df)
plt.title('Passenger Class vs Survival')
plt.xlabel('Passenger Class')
plt.ylabel('survived Count')
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title('Gender vs Survival')
plt.xlabel('Gender')
plt.ylabel('Survived Count')
plt.show()
```

```python
sns.histplot(data=df, x='age', kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.axvline(df['age'].mean(), color='red', linestyle='dashed', linewidth=1, label='Mean Age')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.distplot(df['age'])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.relplot(x='age', y='fare', col='pclass', hue='survived', style='sex', kind='line', data=df)
plt.show()
```

```python
sns.scatterplot(x="age", y="fare", hue="sex", data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x="sex", y="survived", hue="pclass", data=df)
plt.show()
```

```python
sns.stripplot(x="sex", y="age",hue='sex', data=df)
plt.show()
```

```python
sns.swarmplot(x="sex", y="age",hue='sex', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(x="survived", y="age", data=df)
plt.show()
```

```python
sns.violinplot(x="survived", y="age", hue='sex', data=df)
plt.show()
```

```python
sns.countplot(x="survived", data=df, palette="Blues");
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
sns.pointplot(x="sex", y="survived", hue="pclass", data=df)
plt.show()
```

```python
sns.lmplot(x="age", y="fare", data=df)
plt.show()
```

```python
g = sns.FacetGrid(df, col='survived', row='pclass')

g.map(plt.hist, 'age')
g.add_legend()
plt.show()
```

```python
import plotly.express as px



# 1. Bar Chart - Count of Survived Passengers
fig1 = px.bar(df, x="survived", title="Survival Count", labels={"Survived": "Survived (0 = No, 1 = Yes)"})
fig1.show()
```

```python
# 2. Histogram - Age Distribution
fig2 = px.histogram(df, x="age", nbins=30, title="Age Distribution of Passengers", labels={"Age": "Age"})
fig2.show()
```

```python
# 3. Scatter Plot - Fare vs. Age
fig3 = px.scatter(df, x="age", y="fare", color="survived", title="Fare vs. Age",
                  labels={"Age": "Age", "Fare": "Fare Paid", "Survived": "Survived"})
fig3.show()
```

```python
# 4. Bar Chart - Survival Rate by Class
fig4 = px.bar(df, x="pclass", y="survived", color="pclass", title="Survival Rate by Passenger Class",
              labels={"Pclass": "Passenger Class", "Survived": "Survival Rate"}, barmode="group")
fig4.show()
```

```python
fig = px.histogram(df, x="age", facet_col="survived", facet_row="pclass",
                   title="Age Distribution by Survival and Class",
                   labels={"Age": "Age", "Survived": "Survived", "Pclass": "Passenger Class"},
                   nbins=30, color="survived")

fig.update_layout(height=600, width=800)
fig.show()
```
