```python
import pandas as pd
```

```python
df = pd.read_csv('sample_dataset_1000.csv')
```

```python
df.shape
# df.head()
```

```python
df.isnull().sum()
```

if age and salary columns have any missing value

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
```

if name or gender is missing

```python
#mode() ka output list/series hota hay...
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Name'] = df['Name'].fillna(df['Name'].mode()[0])
```

if joining date is missing

```python
df['Joining Date'] = df['Joining Date'].fillna(df['Joining Date'].min())
```

if comments are missing, write your own or write unknown

**Label Transform for categorical data**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['Gender'] = le.fit_transform(df['Gender'])
df['Is Active'] = le.fit_transform(df['Is Active'])
```

```python
df.head()
```

**one hot encoding**

```python
encoded_col = pd.get_dummies(df['Is Active'], prefix='IsActive')

df = pd.concat([df, encoded_col], axis=1).drop(columns=['Is Active'])

df.to_csv('cleaned_sample_1000.csv', index=False)

df.head()
```
