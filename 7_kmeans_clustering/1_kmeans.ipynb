```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
dataset = load_iris()
```

```python
X = dataset.data
df = pd.DataFrame(data=X, columns=dataset.feature_names)

y_true = dataset.target
df['Actual_label'] = y_true
```

```python
model = KMeans(n_clusters=3, random_state=42)
df['Cluster_Label'] = model.fit_predict(X)
```

```python
sns.pairplot(df, hue='Cluster_Label', palette='viridis', corner=True)
plt.suptitle('K-Means Clustering on Iris Dataset', y=1.02)
plt.show()
```

```python
# visualize clustering in 2D using first two features
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=df['Cluster_Label'], palette='viridis', s=100)
centers = model.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.xlabel(dataset.feature_names[0])
plt.ylabel(dataset.feature_names[1])
plt.title('K-Means Clustering (2 Features)')
plt.legend()
plt.grid(True)
plt.show()
```
